```python
import time
from datetime import datetime, date, time, timedelta

GetDriveInformationCount = 0
GDI_counts = []
GDI_datetime = []
PostDriveStatusCount = 0
PDS_counts = []
PDS_datetime = []
PUTManagedDiskCount = 0
PMD_counts = []
PMD_datetime = []
date_time_str_initial = smallDF['DateTime'][0]
date_time_obj_initial = datetime.strptime(date_time_str_initial, '%d/%m/%Y %H:%M:%S')
#GDI_datetime.append(date_time_obj_initial)
#PDS_datetime.append(date_time_obj_initial)
#PMD_datetime.append(date_time_obj_initial) 
ind = 0
ind2 = 0
while(ind < len(smallDF.index)):
    ind2 = ind2 + 1
    GetDriveInformationCount = 0
    PostDriveStatusCount = 0
    PUTManagedDiskCount = 0
    OperationName = smallDF['Operation'][ind]
    date_time_str = smallDF['DateTime'][ind]
    date_time_obj = datetime.strptime(date_time_str, '%d/%m/%Y %H:%M:%S')
    while(date_time_obj < date_time_obj_initial + timedelta(minutes = ind2) and ind < len(smallDF.index)):
        OperationName = smallDF['Operation'][ind]
        date_time_str = smallDF['DateTime'][ind]
        date_time_obj = datetime.strptime(date_time_str, '%d/%m/%Y %H:%M:%S')
        if OperationName == 'GetDriveInformation':
            GetDriveInformationCount = GetDriveInformationCount + 1
            
        elif OperationName == 'PostDriveStatus':
            PostDriveStatusCount = PostDriveStatusCount + 1
            
        elif OperationName == 'PUTManagedDisk':
            PUTManagedDiskCount = PUTManagedDiskCount + 1
        
        ind = ind + 1
        
    GDI_datetime.append(date_time_obj_initial + timedelta(minutes = ind2))
    GDI_counts.append(GetDriveInformationCount)
    PDS_datetime.append(date_time_obj_initial + timedelta(minutes = ind2))
    PDS_counts.append(PostDriveStatusCount)
    PMD_datetime.append(date_time_obj_initial + timedelta(minutes = ind2))
    PMD_counts.append(PUTManagedDiskCount)
    ind = ind + 1
        
```
